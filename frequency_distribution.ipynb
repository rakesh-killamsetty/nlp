```python
import nltk
nltk.download('punkt')
from nltk.probability import FreqDist
```

```python
text11="NLTK is a leading platform for building python programs to work with human language data NLTK is a"
tokens=nltk.word_tokenize(text11)

fdist=FreqDist(tokens)

total_counts=fdist.N()
freq_counts=fdist.B()

text_length=len(text11)

print(total_counts)
print(freq_counts)
print(text_length)
fdist.plot()
```

```python
most_common_words=fdist.most_common()
```

```python
for item,freq in most_common_words:
  print(item,freq)
```
